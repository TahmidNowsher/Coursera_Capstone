```python
import pandas as pd
```

```python
# The url of the wikipedia page
url = 'https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M'
```

```python
df = pd.read_html(url, header = 0)
df
```

```python
# Our desired table is in the first list of df
df1 = pd.DataFrame(df[0])
df1
```

```python
df1.shape
```

```python
df1.drop(df1[df1['Borough'] == 'Not assigned'].index, inplace = True) 
df1
```

```python
#Now we'll reset the inde
df1 = df1.reset_index(drop = True)
df1
```

```python
df1.shape
```

```python
df2 = pd.read_csv('http://cocl.us/Geospatial_data')
```

```python
df2.shape
```

```python
df1.insert(3,'Latitude','')
df1.insert(4,'Longitude','')
```

```python
df1 = df1.set_index('Postal Code')
```

```python
df2 = df2.set_index('Postal Code')
```

```python
index1 = list(df1.index)
```

```python
for x in index1:
    df1.loc[x, 'Latitude'] = df2.loc[x, 'Latitude']
    df1.loc[x, 'Longitude'] = df2.loc[x, 'Longitude']
```

```python
df1 = df1.reset_index()
```
